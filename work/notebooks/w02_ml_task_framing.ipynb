## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

The goal is to rank webpages based on their need for a content refresh. Each webpage receives a priority score so that SEO teams know which pages should be updated first. This is a ranking problem because we care more about the order of pages than simply assigning categories.

```python
print("ML Task Type: Ranking (Scoring)")
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The dataset does not contain a direct "refresh priority" label, so a proxy target is used.

The proxy is based on search performance indicators such as clicks, impressions, CTR, and average position. Pages showing weaker performance or declining trends would receive a higher refresh priority.

This proxy represents the decision that an SEO team would make after reviewing search performance.

```python
import pandas as pd
import numpy as np

# Load your dataset (change path if needed)
df = pd.read_csv("/content/content_refresh_anonymized.csv")

# Example proxy target for demonstration
df["refresh_priority"] = np.random.choice(
    ["High", "Medium", "Low"],
    size=len(df)
)

df[["refresh_priority"]].head()
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The main success metric is **NDCG (Normalized Discounted Cumulative Gain)**.

NDCG measures how well the model ranks webpages according to their true refresh priority.

A higher NDCG value means the most important pages appear near the top of the ranking.

If this problem is later converted into classification, F1-score can also be used.

```python
print("Primary Success Metric: NDCG")
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

One row represents one webpage (URL).

Each webpage contains search performance features such as clicks, impressions, CTR, and average position.

The model predicts the refresh priority for each webpage.

```python
print("Dataset Shape:", df.shape)

df.head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule like "refresh pages with fewer than 500 clicks" ignores many important factors.

Search performance depends on multiple variables, including clicks, impressions, CTR, average position, and their interactions.

Machine learning can learn patterns from several features at the same time and rank webpages more accurately than a fixed rule.

This helps content teams focus on pages with the highest potential impact.
